Part A: Translation, Conceptual

```R
x <- c(1, 3, 5, 7)
y <- ifelse(x > 4, x^2, x + 1)
mean(y)
```

1. Predict Output:
mean y would be 20 in this case. R uses the if else by testing which numbers in x are greater than 4, for the numbers that are they are squared, if they are not greater than 4 they get 1 added instead. y in that case would be c(2,4,25,49). mean(y) calculates the average (2+4+25+49)/4 = 20.

2. Translate to Python:

I used numpy to easily apply a similar if/else to an array.

```Python
import numpy as np

x = np.array([1, 3, 5, 7])

y = np.where(x > 4, x**2, x + 1)

np.mean(y)
```

3. Compute result in python:


````python
import numpy as np

x = np.array([1, 3, 5, 7])
y = np.where(x > 4, x**2, x + 1)
np.mean(y)
````

Part B: Lists vs NumPy Arrays

1. Create a Python list of numbers

````python
a = [1,2,3,4,5]
a
````

2. Attempt to multiply the list by 2:


````python
a*2
````

3. Convert the list to a NumPy array and repeat the multiplication.

````python
#numpy already imported because of Part A.
array_a=np.array(a)
array_a*2
````

**What happened in each case?**

In the case of the list, python does not automatically understand the list is composed of values that can be operated on. The *2 operation is applied to the list as a single element, so you get it repeated.

In the case of the array, we essentially tell python that each of the elements is a numeric value that should be operated on, so we get something similar to what we'd get if we were doing the operation in R, where each of the elements of the array is multiplied by 2.

**Why does NumPy behave differently?**

NumPy interprets it as a numeric array.
